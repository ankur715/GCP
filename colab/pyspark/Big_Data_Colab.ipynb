```python
df = spark.read.csv('housing (1).csv', inferSchema=True, header=True)
df.head(5)
```

```python
df.printSchema()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression

# input all the features in one vector column
assembler = VectorAssembler(inputCols=["crim","zn","indus","chas","nox","rm","age",
                                       "dis","rad","tax","ptratio","b","lstat","medv"],
                            outputCol="Attributes")
output = assembler.transform(df)
```

```python
output
```

```python
# input vs output
finalized = output.select("Attributes","medv")
finalized.show()
```

```python
# split train and test
train, test = finalized.randomSplit([0.8,0.2])
```

```python
regressor = LinearRegression(featuresCol="Attributes", labelCol="medv")
regressor = regressor.fit(train)
pred = regressor.evaluate(test)
pred.predictions.show()
```

```python
# coefficients of regression model
coef = regressor.coefficients

#X and y intercept
intr = regressor.intercept
```

```python
print("Coefficients:", coef)
print("Intercept:", intr)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
eval = RegressionEvaluator(labelCol="medv", predictionCol="prediction", metricName="rmse")
```

```python
# root mean square error
rmse = eval.evaluate(pred.predictions)
print("RMSE:", rmse)

# mean square error
mse = eval.evaluate(pred.predictions, {eval.metricName: "mse"})
print("MSE:", mse)

# mean absolute error
mae = eval.evaluate(pred.predictions, {eval.metricName: "mae"})
print("MAE:", mae)

# r2 - coefficient of determination
r2 = eval.evaluate(pred.predictions, {eval.metricName: "r2"})
print("R2:", r2)
```
